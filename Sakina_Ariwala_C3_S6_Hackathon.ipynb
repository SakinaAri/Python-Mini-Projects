```python
#Making inventories
rice = 30
oats = 20
bread = 10
chicken = 20
milk = 20
oil = 20

# cost per unit of inventories
cost_rice = 100
cost_oats = 60
cost_bread = 20
cost_chicken = 100
cost_milk = 50
cost_oil = 90

```

```python
def add_to_basket():
    print('-------------------Essential items available--------------------------')
    print('%10s  %12s  %10s  %12s' %('S.no','Items','Quantity','Price/unit'))
    print('%10d  %12s  %10d  %12.2f' %(1,'rice',rice,cost_rice))
    print('%10d  %12s  %10d  %12.2f' %(2,'oats',oats,cost_oats))
    print('%10d  %12s  %10d  %12.2f' %(3,'bread',bread,cost_bread))
    print('%10d  %12s  %10d  %12.2f' %(4,'chicken',chicken,cost_chicken))
    print('%10d  %12s  %10d  %12.2f' %(5,'milk',milk,cost_milk))
    print('%10d  %12s  %10d  %12.2f' %(6,'oil',oil,cost_oil))
    
    
    qty_rice = int(input('Enter the quantity of rice: '))
    if qty_rice>rice:
        print('Sorry :( Item out of stock,keep shopping with the available items')
        qty_rice=0
    qty_oats = int(input('Enter the quantity of oats: '))
    if qty_oats>oats:
        print('Sorry :( Item out of stock,keep shopping with the available items')
        qty_oats=0
    
    qty_bread = int(input('Enter the quantity of bread: '))
    if qty_bread>bread:
        print('Sorry :( Item out of stock,keep shopping with the available items')
        qty_bread=0
    
    qty_chicken = int(input('Enter the quantity of chicken: '))
    if qty_chicken>chicken:
        print('Sorry :( Item out of stock,keep shopping with the available items')
        qty_chicken=0
    
    qty_milk = int(input('Enter the quantity of milk: '))
    if qty_milk>milk:
        print('Sorry :( Item out of stock,keep shopping with the available items')
        qty_milk=0
    
    qty_oil = int(input('Enter the quantity of oil: '))
    if qty_oil>oil:
        print('Sorry :( Item out of stock,keep shopping with the available items')
        qty_oil=0
    
    return qty_rice,qty_oats,qty_bread,qty_chicken,qty_milk,qty_oil

def total_item_cost(qty_rice,qty_oats,qty_bread,qty_chicken,qty_milk,qty_oil):
    
    total_cost_items = qty_rice*cost_rice +qty_oats*cost_oats+ qty_bread *cost_bread+ qty_chicken*cost_chicken + qty_milk*cost_milk+ qty_oil*cost_oil
    return total_cost_items

def delivery_details():
    print('Kindly enter your delivery details')
    name = input('Enter your name: ')
    address = input('Enter your address')
    distance = int(input('Enter your distance from the store'))
    return name,address,distance

def delivery_charges(distance):
    if distance <=3:
        delivery_charge=0
    elif distance <=7:
        delivery_charge=50
    else:
        delivery_charge=100
        
    return delivery_charge

def final_bill(total_cost_items,delivery_charge):
    gross_bill = total_cost_items+delivery_charge
    return gross_bill

def generate_bill(qty_rice,qty_oats,qty_bread,qty_chicken,qty_milk,qty_oil,total_cost_items,gross_bill,name,address):
    print('\033[1;3m         🍇 Welcome to Berries & Blooms Grocery 🍇            \033[0m')
    print('                     YOUR BILLING AMOUNT\n                              ')
    print('%10s  %12s  %10s  %12s' %('S.no','Items','Quantity','Price/unit'))
    print('%10d  %12s  %10s  %12.2f' %(1,'rice',rice,cost_rice))
    print('%10d  %12s  %10s  %12.2f' %(2,'oats',oats,cost_oats))
    print('%10d  %12s  %10s  %12.2f' %(3,'bread',bread,cost_bread))
    print('%10d  %12s  %10s  %12.2f' %(4,'chicken',chicken,cost_chicken))
    print('%10d  %12s  %10s  %12.2f' %(5,'milk',milk,cost_milk))
    print('%10d  %12s  %10s  %12.2f' %(6,'oil',oil,cost_oil))
    
    print('Total cost of Purchased  Items : ',total_cost_items)
    print('Total Bill Amount :- Total cost of items + delivery charge', gross_bill)
    print('Your name : ', name)
    print('Your address : ', address)
    
    
while True:
    qty_rice,qty_oats,qty_bread,qty_chicken,qty_milk,qty_oil=add_to_basket()
    total_cost_items=total_item_cost(qty_rice,qty_oats,qty_bread,qty_chicken,qty_milk,qty_oil)
    rice=rice-qty_rice
    oats=oats-qty_oats
    bread=bread-qty_bread
    chicken=chicken-qty_chicken
    milk=milk-qty_milk
    oil=oil-qty_oil
    name,address,distance = delivery_details()
    delivery_charge=delivery_charges(distance)
    gross_bill=final_bill(total_cost_items,delivery_charge)
    generate_bill(qty_rice,qty_oats,qty_bread,qty_chicken,qty_milk,qty_oil,total_cost_items,gross_bill,name,address)
    choice = input('Would you like to keep shopping with us? (yes/no)')
    if choice=='yes':
        continue
    else:
        print('-----------------------------Thankyou for shopping with us! Have a great day.----------------------------------------')
        break       
```
